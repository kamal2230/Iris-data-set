```python
com=PlaintextCorpusReader("D:/Datasets/mini_newsgroups/comp.os.ms-windows.misc",".*")
```

```python
def token(text):
    text=re.sub(r'\W+|\d+|_', ' ', text)
    text=nltk.word_tokenize(text)
    text=[x for x in text if x not in stop]
    text=[l.lemmatize(x) for x in text]
    return text
```

```python
p=[(token(politics.raw(i)) , "politics") for i in politics.fileids()]
```

```python
p
```

```python
c=[(token(com.raw(i)),"computer") for i in com.fileids()]
```

```python
c
```

```python
document = c+p
```

```python
import random
random.seed(50)
random.shuffle(document)
```

```python
lis=[]
for t in document:
    for w in t[0]:
        lis.append(w)
```

```python
def ff(docs):
    w=set(docs)
    f={}
    for x in lis:
        f[x]=(x in w)
    return f
```

```python
fset=[(ff(rev), category) for(rev,category) in document]
```

```python
train=fset[:160]
test=fset[160:]
```

```python
c=nltk.NaiveBayesClassifier.train(train)
```

```python
nltk.classify.accuracy(c,test)
```
